```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
MSE=mean_squared_error(acutal_sales,model_predict_sales)
MSE
```

```python
MAE=mean_absolute_error(acutal_sales,model_predict_sales)
MAE
```

```python
r2 = r2_score(acutal_sales,model_predict_sales)
r2
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
mean_absolute_percentage_error(acutal_sales,model_predict_sales)
```
